```python
!pip install gcsfs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import os
import gcsfs
import tensorflow as tf
from tensorflow.python.lib.io import file_io
```

```python
#Approach 1
# path =r'C:/Users/depri/Desktop/OPS813 cloud computing/Project/case1/amex-nyse-nasdaq-stock-histories/fh_20190217/full_history'
fs = gcsfs.GCSFileSystem(project='Case1-TTPRSD')
# allFiles = glob.glob(path + "/*.csv")
allFiles = fs.ls('stockdatagrp21/full_history')
# allFiles
alldf = []
i=0

for file_ in allFiles:
    i+=1
    if i==8000:
        break
    with fs.open(file_) as f:
      dftemp = pd.read_csv(f,index_col=None, header=0)
#     dftemp = pd.read_csv(file_,index_col=None, header=0)
#     dftemp = df_npd.compute()
    dftemp = dftemp.sort_values('date')
    #Calculating the 200 day moving average before merging all data
    dftemp["200d"] = np.round(dftemp["adjclose"].rolling(window = 200).mean(), 2)
    #Calculating the 50 day moving average before merging all data
    dftemp["50d"] = np.round(dftemp["adjclose"].rolling(window = 50).mean(), 2)
    #Adding a new colmn stock name to identify the stock in question
    dftemp["StockName"]=os.path.basename(file_).split('.')[0]
    alldf.append(dftemp)

df = pd.concat(alldf, axis = 0, ignore_index = True)
df.columns = ["Date", "Volume","Open", "Close", "High", "Low", "AdjustedClose","200d","50d","StockName"]
# print(df) 


#Approach 2
# def read_csv_file(filename):
#   with file_io.FileIO(filename, 'r') as f:
#     dftemp = pd.read_csv(f)
#     dftemp = dftemp.sort_values("date")
#     #Calculating the 200 day moving average before merging all data
#     dftemp["200d"] = np.round(dftemp["adjclose"].rolling(window = 200).mean(), 2)
#     #Calculating the 50 day moving average before merging all data
#     dftemp["50d"] = np.round(dftemp["adjclose"].rolling(window = 50).mean(), 2)
#     #Adding a new colmn stock name to identify the stock in question
#     dftemp["StockName"]=os.path.basename(filename).split('.')[0]
#   return dftemp
# def read_csv_files(filename_pattern):
#   filenames = tf.gfile.Glob(filename_pattern)
#   dataframes = [read_csv_file(filename) for filename in filenames]
#   return pd.concat(dataframes)

# df = read_csv_files(os.path.join('gs://stockdatagrp2/full_history', '*.csv'))
# df.columns = ["Date", "Volume","Open", "Close", "High", "Low", "AdjustedClose","200d","50d","StockName"]
```

```python
# (1) find all stocks that are currently below their 200 day moving average.

#Dropping all rows with 200 day moving average as NaN
dfna = df.dropna()
#Identifying most recent date per stock
dfdate = dfna.groupby('StockName').max()['Date']
# dfdate
keys = dfdate.index
# keys
#Filtering all recent stock who is currently below their 200 day moving average
# dfna.loc[(dfna['AdjustedClose']<dfna['200d'])&(dfna['Date']=='2019-02-15')]
below200=dfna.loc[(dfna['AdjustedClose']<dfna['200d'])&(dfna['StockName'].isin(keys))&(dfna['Date'].isin(dfdate.loc[keys]))]
below200.head(7)
```

```python
# (2) find all stocks that are currently above their 200 day moving average
# dfna.loc[(dfna['AdjustedClose']>dfna['200d'])&(dfna['Date']=='2019-02-15')]
above200=dfna.loc[(dfna['AdjustedClose']>dfna['200d'])&(dfna['StockName'].isin(keys))&(dfna['Date'].isin(dfdate.loc[keys]))]
above200.head(7)
```

```python
#Finding stocks in the last month with golden cross and visualizing it
trendplot = df[df.Date > '2019-01-15'][['Date','AdjustedClose','200d','50d','StockName']]
trendplot['50d-200d'] = trendplot['50d'] - trendplot['200d']
trendplot['Trend'] = np.where(trendplot['50d-200d'] > 0,1,0)
trendplot['Trend'] = np.where(trendplot['50d-200d'] < 0,1,trendplot['Trend'])
# Debugging plot for 
for key, grp in trendplot.groupby('StockName'):
  previous_50 = grp['50d'].shift(1)
  previous_200 = grp['200d'].shift(1)
  crossing = (((grp['50d'] <= grp['200d']) & (previous_50 >= previous_200))
              | ((grp['50d'] >= grp['200d']) & (previous_50 <= previous_200)))
  crossing = crossing.loc[crossing]
  if crossing.empty:
    continue
  fig, ax = plt.subplots()
  fig.suptitle(key)
  ax = grp.plot(ax=ax, kind='line', x='Date', y=['AdjustedClose','200d','50d'], label=key,figsize=(20,10))

plt.legend(loc='best');
plt.show();
```
